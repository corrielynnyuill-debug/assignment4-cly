```python
# Task 3: Handling Null Values
# Create a new DataFrame with some null values and fill them

# Your code here:
df_null = pd.DataFrame({
    "Name": ["Alice", "Bob", None, "David"],
    "Age": [25, 30, None, 28],
    "City": ["New York", "San Francisco", "Los Angeles", None]
})

print("DataFrame with Null Values:\n")
print(df_null)
print("\nNull Value Count:\n")
print(df_null.isnull().sum())

try:
  df_filled = df_null.copy()
  df_filled["Name"] = df_filled["Name"].fillna("Unknown")
  df_filled["Age"] = df_filled["Age"].fillna(df_filled["Age"].mean()).astype(int)
  df_filled["City"] = df_filled["City"].fillna("Unknown")

  print("\nDataFrame after Filling Null Values:\n")
  print(df_filled)
except Exception as e:
  print(f"An error occurred: {e}")

```

```python
# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:

print("Iterator for first 5 even numbers:")

try:
  # Iterator: using inter() on a list comprhension
  even_list = [2 * i for i in range (1,6)]
  even_iterator = iter(even_list)

  for num in even_iterator:
    print(num)

except Exception as e:
  print(f"An error occured: {e}")

print("\nGenerator for first 5 even numbers:")

try:
  # Generator: using yeild
  def even_generator(n):
    for i in range(1, n, + 1):
      yield 2 * i

  for num in even_generator(5):
    print(num)
except Exception as e:
  print(f"An error occured: {e}")
```

```python
# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

numbers = [1, 2, 3, 4, 5]

# Your code here:
from functools import reduce

print("Original List:", numbers)

try:
  # Map: square all numbers
  squared_numbers = list(map(lambda x: x * x, numbers))
  print("\nSquared Numbers using Map:", squared_numbers)

  # Reduce: Product of all numbers
  product = reduce(lambda x, y: x * y, numbers)
  print("\nProduct of all numbers using Reduce:", product)

  # Filter: Get only even numbers
  even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
  print("\nEven Numbers using Filter:", even_numbers)

except Exception as e:
  print(f"An error occured: {e}")
```

```python
# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:
class Rectangle:
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def area(self):
    """Calculate ad return the area of the rectangle."""
    return self.length * self.width

  def perimeter(self):
    """ Calculate and return the perimeter of the rectangle."""
    return 2 * (self.length + self.width)


# Demonstrate
try:
  rect = Rectangle(10,5)
  print("Rectangle Details:")
  print("Length:", rect.length)
  print("Width:", rect.width)
  print("Area:", rect.area())
  print("Perimeter:", rect.perimeter())
except Exception as e:
  print(f"An error occured: {e}")
```

```python
# Task 12: Pandas Data Analysis
# Using the following DataFrame, perform these tasks:
# 1. Find the average salary by department
# 2. Get the names of employees with salary > 60000
# 3. Add a new column 'Bonus' which is 10% of salary

df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# Your code here:
try:
  # Average Salary by department
  avg_salary = df_employees.groupby("Department")["Salary"].mean()
  print("Average Salary by Department:\n")
  print(avg_salary)

  # Employees with a salary higher than 60K
  high_salary_employees = df_employees[df_employees["Salary"]>60000]["Name"]
  print("\nEmployees with Salary > 60K:\n")
  print(high_salary_employees)

  # Bonus Column 10% of Salary
  df_employees["Bonus"] = df_employees["Salary"] * 0.1
  print("\nDataFrame with Bonus Column:\n")
  print(df_employees)

except Exception as e:
  print(f"An error occured: {e}")
```
